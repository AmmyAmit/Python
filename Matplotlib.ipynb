# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0,10])
ypoints = np.array([5,50])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
x = np.array([5,20])
y = np.array([10,120])
plt.plot(x,y)
plt.show()
```

```python
p=np.array([1,0.5])
q=np.array([3,2])
plt.plot(p,q)
plt.show()
```

```python
plt.plot(x,y,'*')
plt.show()
```

```python
plt.plot(p,q,'+')
plt.show()
```

```python
ypoints= np.array([1,5,10,20])
plt.plot(ypoints,marker='o')
plt.show()
```

```python
xpoints= np.array([1,5,10,20])
plt.plot(xpoints,marker='o')
plt.show()
```

```python
y1 = np.array([5,6,1,10])
y2 = np.array([6,12,45,98])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1 = np.array([1,5,6,7,9,15])
x2 = np.array([2,6,4,3,5,9])
y1 = np.array([3,6,8,7,9,6])
y2 = np.array([1,6,3,2,8,10])
plt.plot(x1,x2,y1,y2)
plt.xlabel('year')
plt.ylabel('petrol price')
plt.title('variation in price of petrol')
plt.show()
```

```python
zpoints = np.array([1,5,9,11,15])
plt.plot(zpoints,marker='*',linestyle='dashed',mec='red')
plt.show()
```
